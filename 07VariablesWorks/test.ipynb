```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

dbscan = DBSCAN(eps=0.2, min_samples=1)
clusters = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o', edgecolor='k')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
def func(str1, str2):
    dp = [[0] * len(str1) for _ in range(len(str2))]
    
    count = 0
    match = False
    
    for i in range(len(str2)):
        for j in range(len(str1)):
            if i == 0:
                if str1[j] == str2[i] and count == 0:
                    dp[i][j] = 1
                    match = True
                else:
                    dp[i][j] = 0
                    match = match or False
            else:
                if str1[j] == str2[i] and dp[i - 1][j] != 1:
                    dp[i][j] = 1
                    match = True
                else:
                    dp[i][j] = 0
                    match = match or False
        if match:
            count += 1
            if count == len(str1):
                return True
            match = False
        else:
            count = 0
    
    return False

s = "abc"
s2 = "lecaxbee"

func(s, s2)
```
