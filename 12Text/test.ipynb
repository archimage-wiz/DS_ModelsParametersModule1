```python
from spellchecker import SpellChecker

text = """
Обработка текстов на естественном языке — общее направление искусственного интеллекта и математической лингвистики.
Оно изучает проблемы компьютерного анализа и синтеза текстов на естественных языках.
Применительно к искусственному интеллекту анализ означает понимание языка, а синтез — генерацию грамотного текста.
"""

spell = SpellChecker(language='ru')

corrected_words = [spell.correction(word) for word in text.split()]
xx : list = list(filter(lambda word: word != None, corrected_words))
corrected_words = ' '.join(xx)
corrected_words
```

```python
from pymystem3 import Mystem
import re

mystem = Mystem()
```

```python


stop_words = stopwords.words('russian')
sentences = nltk.sent_tokenize(text, language="russian")
for sentence in sentences:
    print(1, sentence)
    sentence_ = re.sub(r"[^А-Яа-яёЁ ]", "", sentence)
    print(2, sentence_)
    tokens = nltk.word_tokenize(sentence_, language="Russian")
    print(3, tokens)
    tokens = [i for i in tokens if (i not in stop_words)]
    print(4, tokens)
    tokens = mystem.lemmatize(' '.join(tokens))[:-1]
    print(5, tokens)
    tokens = list(filter(lambda word: word != ' ', tokens))
    print(6, tokens)
    print()
```

```python
from natasha import (
    Segmenter,
    MorphVocab,
    NewsEmbedding,
    NewsMorphTagger,
    Doc
)

# Инициализация инструментов
segmenter = Segmenter()
morph_vocab = MorphVocab()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)

# Текст для анализа
# text = "кошки гуляли по крыше"

# Создание документа
doc = Doc(text)
display(doc)
# Сегментация и морфологический анализ
doc.segment(segmenter)
display(doc.tokens)
doc.tag_morph(morph_tagger)
display(doc.morph)

# Лемматизация
for token in doc.tokens:
    token.lemmatize(morph_vocab)
    print(token.text, '->', token.lemma)
```
