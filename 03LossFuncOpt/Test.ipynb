```python
import numpy as np

# Создаем данные
# np.random.seed(42)
# X = 2 * np.random.rand(100, 1)
# y = 4 + 3 * X + np.random.randn(100, 1)
X = 2 * np.random.rand(100, 1)
y = 20 - X*2 + np.random.randn(100, 1)

# Инициализация параметров
w = np.random.randn(1)
b = np.random.randn(1)
# Гиперпараметры
learning_rate = 0.002
n_iterations = 5000

# Функция для вычисления предсказаний
def predict(X, w, b):
    return X.dot(w) + b

# Основной цикл SGD
for iteration in range(n_iterations):
    idx = np.random.randint(len(X))
    X_i = X[idx:idx+1]
    y_i = y[idx:idx+1]
    
    # Вычисляем предсказания
    y_pred = predict(X_i, w, b)
    
    # Вычисляем градиенты
    gradient_w = 2 * X_i.dot(y_i - y_pred)
    gradient_b = 2 * (y_i - y_pred)
    
    # Обновляем параметры
    w = w + learning_rate * gradient_w
    b = b + learning_rate * gradient_b
    
    if iteration % 100 == 0:
        loss = np.mean((y - predict(X, w, b)) ** 2)
        print(f"Iteration {iteration}, Loss: {loss}")

print(f"Final parameters: w = {w}, b = {b}")
```

```python
from sklearn.linear_model import LinearRegression

X = 2 * np.random.rand(100, 1)
y = 20 - X*2 + np.random.randn(100, 1)

model = LinearRegression()
model.fit(X, y)
display(model.score(X, y))

plt.plot(X, y, "b.")
plt.plot(X, model.predict(X), "r-")
print(model.intercept_, model.coef_)
model.intercept_ = np.array([19.42397462])
model.coef_ = np.array([[-1.37085765]])
print(model.intercept_, model.coef_)
plt.plot(X, model.predict(X), "g-")
plt.show()
```
